```python
import pandas as pd
```

```python
data = pd.read_csv('/content/코스피_2020.csv')
```

```python
data
```

```python
filtered_df_1 = data[~data['1'].str.contains('면')]
filtered_df_1
```

```python
filtered_df_1.rename(columns={'1':'Date'}, inplace=True)
filtered_df_1.rename(columns={'0':'Title'}, inplace=True)
filtered_df_1
```

```python
filtered_df_1['Date'] = filtered_df_1['Date'].str.replace(r'\.$', '')
```

```python
filtered_df_1['Date'] = filtered_df_1['Date'].apply(lambda x: pd.to_datetime(str(x), format='%Y-%m-%d'))
```

```python
filtered_df_1
```

```python
filtered_df_1.reset_index(drop=True)
```

```python
df = filtered_df_1
df
```

```python
import os
import torch

# Allocate GPU in Colab
os.environ["CUDA_DEVICE_ORDER"] = "PCI_BUS_ID"
os.environ["CUDA_VISIBLE_DEVICES"] = "0"  # You can adjust the GPU index as needed

device = "cuda:0" if torch.cuda.is_available() else "cpu"
```

```python
texts = list(df['Title'])
texts[:4]
```

```python
!pip install transformers
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
```

```python
tokenizer = AutoTokenizer.from_pretrained("snunlp/KR-FinBert-SC")
model = AutoModelForSequenceClassification.from_pretrained("snunlp/KR-FinBert-SC").to(device)
```

```python
tokenizer.tokenize(texts[0])
```

```python
pip install tqdm
```

```python
from tqdm import tqdm
```

```python
outputs_list = []
for text in tqdm(texts):
    inputs = tokenizer(text, return_tensors='pt').to(device)
    output = model(**inputs)
    output = output.logits.tolist()[0]
    outputs_list.append(output)
```

```python
outputs = torch.tensor(outputs_list)
outputs
```

```python
import torch.nn as nn
```

```python
predictions = nn.functional.softmax(outputs, dim=-1)
predictions
```

```python
df_sc = pd.DataFrame(predictions.detach().numpy())
df_sc.columns = ['부정','중립','긍정']
df_sc
```

```python
df = df.reset_index(drop=True)
```

```python
df
```

```python
data = pd.concat([df, df_sc], axis=1)
data
```

```python
data.to_csv('K-finbert 성공3.csv', index=False)
```
